### Transformasi data label menjadi numeris

```python
data['DC201']=data['DC201'].apply(lambda x: 0 if x=='Layak Minum' else 1)
data['DC201'].value_counts()
```

### Membuat fungsi untuk mengisi data yang kosong disetiap kolom menggunakan mean/median/modus tiap kolom

```python
def fillData(data:pd.Series,method:str):
    if method=='mean':
        mean=data.mean()
        data=data.fillna(mean)
    elif method=='median':
        median=data.median()
        data.fillna(median)
    else:
        mode=data.mode()[0]
        data.fillna(mode,inplace=True)
    return data
```

### Test memberi nilai pada data kolom yang kosong

```python
import random
random.seed(42)

# Get random index
rand_idx=random.randint(0,len(null_column))

# Get column and jumlah null
column,null=null_column[rand_idx]
method='mode'
# apply function to null column
temp_data=fillData(data=data[column],method=method)
print(f'Null column {column} before filling : {null}\nNull column {column} after filling  {method} : {temp_data.isnull().sum()}\n')
```

```python
null_column
```

### Isi semua data null yang ada di dataset dengan metode yang diinginkan

```python
# Iterasi column yang memiliki null value
method='mode'
for item in null_column:
    column,null=item
    # apply function to null column
    data[column]=fillData(data=data[column],method=method)
    print(f'Null column {column} before filling : {null}\nNull column {column} after filling  {method} : {data[column].isnull().sum()}\n')
```
