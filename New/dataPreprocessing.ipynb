### Fill the null value in each attribute

```python
def fillData(data:pd.Series,method:str):
    if method=='mean':
        mean=data.mean()
        data=data.fillna(mean)
    elif method=='median':
        median=data.median()
        data.fillna(median)
    else:
        mode=data.mode()[0]
        data.fillna(mode,inplace=True)
    return data
```

### Get all null value in attribute

```python
# Check Nan/Null Value and then store it in list of tuple
null=data.isnull().sum().to_dict()
null_column=[(key,value) for key,value in null.items() if value>0]
null_column
```

```python
data['DC205'].value_counts()
```

### Test sample of filling null atribute

```python
import random
random.seed(42)

# Get random index
rand_idx=random.randint(0,len(null_column))

# Get column and jumlah null
column,null=null_column[rand_idx]
method='mode'
# apply function to null column
temp_data=fillData(data=data[column],method=method)
print(f'Null column {column} before filling : {null}\nNull column {column} after filling  {method} : {temp_data.isnull().sum()}\n')
```

### Apply function to all attribute

```python
# Iterasi column yang memiliki null value
method='mode'
for item in null_column:
    column,null=item
    # apply function to null column
    data[column]=fillData(data=data[column],method=method)
    print(f'Null column {column} before filling : {null}\nNull column {column} after filling  {method} : {data[column].isnull().sum()}\n')
```

### The target data are balanced. The next step is finding the relevant feature with some technique:
- Correlation
- Feature Importance (XGBOOST)
- PCA (Principal Component Analysis)

```python
import xgboost as xgb
def selectAttribute(data:pd.DataFrame,method:str,threshold:float=0.0):
    if method=='xgb':
        X=data.drop('DC201',axis=1)
        y=data['DC201']
        # Buat instance model XGBoost
        model = xgb.XGBClassifier()
        # Latih model dengan data
        model.fit(X, y) 
        # Dapatkan skor pentingnya fitur
        importance = model.feature_importances_
        column=data.columns[:-1]
        results={}
        i=0
        for col in column:
            results[col]=importance[i]
            i+=1
        return results
    
    else :
        #Correlation with output variable
        corr=data.corr()
        cor_target = corr["DC201"]
        #  Selecting highly correlated features
        relevant_features = cor_target[cor_target>0]
        selected=relevant_features.keys()
        print(relevant_features)
        return selected
```

```python
results=selectAttribute(data=data,method='xgb',threshold=0)
# Sort the dictionary by value
results = dict(sorted(results.items(), key=lambda x: x[1],reverse=True))
results
```
