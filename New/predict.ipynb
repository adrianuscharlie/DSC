```python
import pandas as pd
import xgboost as xgb
```

```python
test=pd.read_csv('../data/test.csv').astype(float)
test=test.drop('id',axis=1)
test
```

```python
test=pd.read_csv('test_scaled.csv').astype(float)
test
```

## Load the model

```python
loaded_model = xgb.XGBClassifier()
loaded_model.load_model("./xgboost_2.json")
loaded_model
```

## Predict the test data

```python
predict=loaded_model.predict(test)
predict
```

## Load the submission template

```python
submission=pd.read_csv('../data/sample_submission.csv')
submission['DC201']=predict
submission
```

```python
submission['DC201']=submission['DC201'].apply(lambda x:'Layak Minum' if x==1 else 'Tidak Layak Minum')
submission
```
