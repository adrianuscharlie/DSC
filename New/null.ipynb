## Discover the null value

```python
# Check Nan/Null Value and then store it in list of tuple
null=data.isnull().sum().to_dict()
null_column=[key for key,value in null.items() if value>0]
```

```python
data[data['DC205'].isnull()]['DC201'].value_counts()
```

```python
def checkNull(data:pd.DataFrame,column:str):
    data_value=data[data[column].isnull()]
    value=data_value['DC201'].value_counts().to_dict()
    modus_layak=data_value['DC201']
    return value,data
```

```python
null_data=[]
for item in null_column:
    temp={'column':item}
    value,data=checkNull(data=data,column=item)
    temp['value']=value
    temp['data']=data
    null_data.append(temp)
```

### Mencari jumlah null di kedua target label agar mengetahui perbandingan null value

```python
for item in null_data:
    print(f"Column {item['column']} with null value {data[item['column']].isnull().sum()} with",item['value'])
```

## Mencari modus di setiap kolom berdasarkan target label

```python
data.loc[data['DC201'] == 'Layak Minum', 'DC230b'].mode().iloc[0]
```

```python
def findMode(data:pd.DataFrame,column:str):
    mode_layak=data.loc[data['DC201'] == 'Layak Minum', column].mode().iloc[0]
    mode_tidak_layak=data.loc[data['DC201'] == 'Tidak Layak Minum', column].mode().iloc[0]
    mode={'column':column,'Layak Minum':mode_layak,'Tidak Layak Minum':mode_tidak_layak}
    print(mode)
    return mode
```

```python
mode=[]
for item in null_column:
    mode.append(findMode(data,item))
```

## Mengisi setiap null value yang ada di setiap atribut berdasarkan modus di atribut tersebut dengan acuan target label

```python
def fillNull(data:pd.DataFrame,column:str,mode:dict):
    print(mode)
    data.loc[(data['DC201'] == 'Layak Minum') & (data[column].isnull()), column] = mode['Layak Minum']
    data.loc[(data['DC201']=='Tidak Layak Minum') & (data[column].isnull()), column] = mode['Tidak Layak Minum']
    return data
```

```python
for col in mode:
    data=fillNull(data=data,column=col['column'],mode=col)
```
